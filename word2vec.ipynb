```python
import os
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

# Make sure NLTK punkt tokenizer is available
import nltk
nltk.download('punkt')

# Directory where your GoT books are located (uploaded via "Add Data")
data_dir = '/kaggle/input/game-of-thrones-books'

# Initialize list to store tokenized sentences
story = []

# Loop through all files in the uploaded directory
for filename in os.listdir(data_dir):
    file_path = os.path.join(data_dir, filename)
    
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
        corpus = f.read()
        raw_sentences = sent_tokenize(corpus)
        for sent in raw_sentences:
            story.append(simple_preprocess(sent))

# Show a preview
print(f"Total tokenized sentences: {len(story)}")
print("Example:", story[0])
```

```python
len(story)
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```
